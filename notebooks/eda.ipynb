```python
# Create a copy of the dataset to avoid modifying the original
df_continuous = df.copy()

# Assign sequential numbering for each Variety (and optionally Grade)
df_continuous['Sequence'] = df_continuous.groupby(['Variety', 'Grade']).cumcount() + 1

# Plot for each Variety and Grade combination
plt.figure(figsize=(14, 8))

# Group the data by Variety and Grade
grouped = df_continuous.groupby(['Variety', 'Grade'])

for (variety, grade), subset in grouped:
    if not subset.empty:  # Ensure the subset isn't empty
        plt.plot(
            subset['Sequence'], 
            subset['Avg Price'], 
            label=f"{variety} - {grade}", 
            linestyle='-'
        )

# Add legend, labels, and title
plt.legend(loc='best', fontsize=8)
plt.xlabel('Observation Number')
plt.ylabel('Average Price')
plt.title('Continuous Trends by Variety and Grade (Using Sequential Numbering)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
case_weights = {
    'FC': 16,  # FC case type has a weight of 16 kg
    'HC': 8    # HC case type has a weight of 8 kg
}

# Function to convert price to per kg
def price_per_kg(row):
    case_type = row['CaseType']
    
    # Handle missing values
    if pd.isna(case_type):
        return None, None, None  # Return None for missing values
    
    # Retrieve weight based on CaseType
    weight = case_weights.get(case_type)
    
    if weight is None:  # Handle cases where case type is not in the dictionary
        return None, None, None
    
    # Calculate per kg prices
    return row['Min Price'] / weight, row['Max Price'] / weight, row['Avg Price'] / weight

# Apply the function to calculate per kg prices
df[['Min Price (per kg)', 'Max Price (per kg)', 'Avg Price (per kg)']] = df.apply(
    lambda row: pd.Series(price_per_kg(row)), axis=1
)

# Show the DataFrame with the calculated per kg prices
print(df)
```

```python
import os
import pandas as pd
varieties = df['Variety'].unique()
grades = df['Grade'].unique()

# Define the output folder
output_folder = r"C:\Users\Aqib Gul\OneDrive\Videos\Price_forecasting_project\data\raw\processed"

# Create the folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Generate datasets for each variety and grade
output_datasets = {}
start_date = df['Date'].min()
end_date = df['Date'].max()
full_date_range = pd.date_range(start=start_date, end=end_date, freq='D')

for variety in varieties:
    for grade in grades:
        # Filter for specific variety and grade
        subset = df[(df['Variety'] == variety) & (df['Grade'] == grade)].copy()
        
        # Set 'Date' as the index for reindexing
        subset.set_index('Date', inplace=True)
        
        # Reindex to full date range and fill missing data with zeros
        subset = subset.reindex(full_date_range)
        subset.index.name = 'Date'
        subset.reset_index(inplace=True)
        
        # Fill missing data
        subset['District'] = subset['District'].ffill()  # Fill forward district information
        subset['Market'] = subset['Market'].ffill()
        subset['Fruit'] = subset['Fruit'].ffill()
        subset['Variety'] = variety
        subset['Grade'] = grade
        subset['Min Price (per kg)'] = subset['Min Price (per kg)'].fillna(0)
        subset['Max Price (per kg)'] = subset['Max Price (per kg)'].fillna(0)
        subset['Avg Price (per kg)'] = subset['Avg Price (per kg)'].fillna(0)
        
        # Add mask column (1 for existing data, 0 for missing data)
        subset['Mask'] = (subset['Min Price (per kg)'] > 0).astype(int)
        
        # Save to dictionary
        output_datasets[f"{variety}_{grade}"] = subset
        
        # Save each dataset as a CSV file in the specified folder
        output_file = os.path.join(output_folder, f"{variety}_{grade}_dataset.csv")
        subset.to_csv(output_file, index=False)
```

```python
# Load the Maharaji datasets
maharaji_a_path = r"C:\Users\Aqib Gul\OneDrive\Videos\Price_forecasting_project\data\raw\processed\Maharaji_A_dataset.csv"
maharaji_b_path = r"C:\Users\Aqib Gul\OneDrive\Videos\Price_forecasting_project\data\raw\processed\Maharaji_B_dataset.csv"

# Read the datasets
maharaji_a = pd.read_csv(maharaji_a_path, parse_dates=["Date"])
maharaji_b = pd.read_csv(maharaji_b_path, parse_dates=["Date"])

# Function to shift data to November-December and handle duplicate dates
def shift_to_november_december(df):
    df['Date'] = df['Date'] + pd.DateOffset(months=1)  # Shift dates by 1 month
    
    # Aggregate to remove duplicates (if any)
    df = df.groupby('Date', as_index=False).agg({
        'District': 'first',
        'Market': 'first',
        'Fruit': 'first',
        'Variety': 'first',
        'Grade': 'first',
        'Min Price (per kg)': 'mean',  # Use mean if multiple rows for the same date
        'Max Price (per kg)': 'mean',
        'Avg Price (per kg)': 'mean',
        'Mask': 'max',  # Keep mask as 1 if any row has data
    })
    
    # Create a complete date range
    full_date_range = pd.date_range(start=df['Date'].min(), end=df['Date'].max(), freq='D')
    
    # Reindex to full date range
    df.set_index('Date', inplace=True)
    df = df.reindex(full_date_range)
    df.index.name = 'Date'
    df.reset_index(inplace=True)
    
    # Fill missing values for non-numeric columns
    df['District'] = df['District'].ffill()
    df['Market'] = df['Market'].ffill()
    df['Fruit'] = df['Fruit'].ffill()
    df['Variety'] = df['Variety'].ffill()
    df['Grade'] = df['Grade'].ffill()
    
    # Fill missing numeric data with zeros
    df['Min Price (per kg)'] = df['Min Price (per kg)'].fillna(0)
    df['Max Price (per kg)'] = df['Max Price (per kg)'].fillna(0)
    df['Avg Price (per kg)'] = df['Avg Price (per kg)'].fillna(0)
    
    # Adjust mask column
    df['Mask'] = (df['Min Price (per kg)'] > 0).astype(int)
    
    return df

# Modify Maharaji datasets for Grade A and Grade B
maharaji_a_modified = shift_to_november_december(maharaji_a)
maharaji_b_modified = shift_to_november_december(maharaji_b)

# Save the modified datasets back to disk
output_folder = r"C:\Users\Aqib Gul\OneDrive\Videos\Price_forecasting_project\data\raw\processed"
maharaji_a_modified.to_csv(f"{output_folder}\\Maharaji_A_modified.csv", index=False)
maharaji_b_modified.to_csv(f"{output_folder}\\Maharaji_B_modified.csv", index=False)

print("Datasets have been successfully modified and saved!")
```
